```python
import pandas as pd
import numpy as np

# Configuración de registros
n_registros = 50000

# 1. Generar datos base
id_aleatorio = np.arange(1, n_registros + 1)

nombres = ['Hunk','Imeri','Joshua','Seah','Israel','Diego']
nombre_aleatorio = np.random.choice(nombres, n_registros)

apellidos = ['Torres','Castillo','Flores','Azul','Mazda','Phoenix']
apellido_aleatorio = np.random.choice(apellidos, n_registros)

genero = ['F', 'M']
gen_aleatorio = np.random.choice(genero, n_registros)

paises = ['España', 'Mexico', 'Guatemala', 'Chile', 'Argentina', 'Perú']
pais_aleatorio = np.random.choice(paises, n_registros)

# --- NUEVOS CAMPOS SEGÚN LA IMAGEN ---

# Estado: soltero(a) o casado(a)
estados = ['Soltero(a)', 'Casado(a)']
estado_aleatorio = np.random.choice(estados, n_registros)

# Salario: entre Q1500 a Q10000
salario_aleatorio = np.random.randint(1500, 10001, size=n_registros)

# Edad: entre 18 a 80 años
edad_aleatoria = np.random.randint(18, 81, size=n_registros)

# Registro: fecha aleatoria en 2026
fechas_2026 = pd.date_range(start='2026-01-01', end='2026-12-31')
registro_aleatorio = np.random.choice(fechas_2026, n_registros)

# --- CREACIÓN DEL DATAFRAME ---

df = pd.DataFrame({
    'id': id_aleatorio,
    'nombre': nombre_aleatorio,
    'apellido': apellido_aleatorio,
    'genero': gen_aleatorio,
    'pais': pais_aleatorio,
    'estado': estado_aleatorio,
    'salario': salario_aleatorio,
    'edad': edad_aleatoria,
    'registro': registro_aleatorio
})

# --- RESOLUCIÓN DE LOS PROBLEMAS PLANTEADOS ---
print(df.head())
print("-" * 30)

print(df.info())
print("-" * 30)

print(df.describe())
print("-" * 30)

print(df['genero'].value_counts())
print("-" * 30)

conteo_pais_genero = df.groupby(['pais', 'genero']).size().unstack()
print(conteo_pais_genero)
```
